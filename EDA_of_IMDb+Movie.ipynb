- ### <font color='blue'>Subtask 1.4 Find the Most Popular Trios -
    
Suppose, We'r looking to make a blockbuster movie. Hence,there will be primarily 3 lead roles in our movie and we wish to cast the most popular actors for it. Now, since we don't want to take a risk. so, we will cast a trio which has already acted in together in a movie before. The metric that we've chosen to check the popularity is the Facebook likes of each of these actors.


```python
# finding the trios which has the most number of Facebook likes combined.
group=Mov.pivot_table(values=["actor_1_facebook_likes","actor_2_facebook_likes","actor_3_facebook_likes"],
                  aggfunc="sum",index=["actor_1_name","actor_2_name","actor_3_name"])

group.head()
```

```python
# Calculating total likes
group["Total likes"]=group["actor_1_facebook_likes"]+group["actor_2_facebook_likes"]+group["actor_3_facebook_likes"]

group.head()
```

```python
#Sorting in descending order Total likes
group.sort_values(by="Total likes",ascending=False,inplace=True)
group
```

```python
#creating index to the group dataframe
group.reset_index(inplace=True)
group
```

```python
# Top 5 popular actors and their popularity
group=group.iloc[0:5,:]
group
```

-  ###  <font color='blue'>Subtask 2.1 Combine the Dataframe by Genres

```python
# Creating a new dataframe `df_by_genre` that contains `genre_1`, `genre_2`, and `genre_3` and all the columns related to CVotes/Votes from the Mov data frame. 
#There are 47 columns to be extracted in total.
df_by_genre=Mov.loc[:,"CVotes10":"VotesnUS"]
df_by_genre[["genre_1","genre_2","genre_3"]]=Mov[["genre_1","genre_2","genre_3"]]
```

```python
df_by_genre.head()
```

```python
# Creating a column cnt and initialize it to 1
df_by_genre["cnt"]=1
df_by_genre.head()
```

```python
df_by_genre[["genre_1","genre_2","genre_3"]]
```

```python
# Grouping the movies by individual genres
#Grouping the dataframe `df_by_genre` by `genre_1` and finding the sum of all the numeric columns such as `cnt`, columns related to CVotes and Votes columns and store it in a dataframe `df_by_g1`.
df_by_g1=df_by_genre.groupby("genre_1").aggregate(np.sum)

#Grouping the dataframe `df_by_genre` by `genre_1` and finding the sum of all the numeric columns such as `cnt`, columns related to CVotes and Votes columns and store it in a dataframe `df_by_g2`.
df_by_g2=df_by_genre.groupby("genre_2").aggregate(np.sum)

##Grouping the dataframe `df_by_genre` by `genre_1` and finding the sum of all the numeric columns such as `cnt`, columns related to CVotes and Votes columns and store it in a dataframe `df_by_g3`.
df_by_g3=df_by_genre.groupby("genre_3").aggregate(np.sum)
```

```python
df_by_g1
```

```python
df_by_g2
```

```python
df_by_g3
```

```python
# Adding the grouped data frames and storing it in a new data frame
df_add=df_by_g1.add(df_by_g2,fill_value=0)
df_add=df_add.add(df_by_g3,fill_value=0)
df_add
```

```python
# Extracting genres with atleast 10 occurences
genre_top_10=df_add.loc[df_add["cnt"]>10]
```

```python
genre_top_10
```

```python
# Taking the mean for every column by dividing with cnt 
genre_top_10.iloc[:,0:-1]=genre_top_10.iloc[:,0:-1].divide(genre_top_10["cnt"],axis=0)
```

```python
genre_top_10
```

```python
# Rounding off the columns of Votes till two decimals
genre_top_10.loc[:,"VotesM":"VotesnUS"]=round(genre_top_10.loc[:,"VotesM":"VotesnUS"],2)
```

```python
# Converting CVotes into int datatype
genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns]=genre_top_10[genre_top_10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
```

```python
genre_top_10
```

```python
# Creating IFUS column in dataframe Mov that contain the value "USA" if the `Country` of the movie is "USA".
Mov['IFUS']=Mov['Country'].copy()

#The column `IFUS` should  For all other countries other than the USA, `IFUS` that contain the value `non-USA`.
Mov.loc[Mov["IFUS"]!="USA","IFUS"]='Non-USA'

Mov['IFUS'].value_counts()
```
